```python
from pathlib import Path
prompt_file_list = list(Path("./prompts").glob("*.md"))
```

```python
result = []
for file_path in prompt_file_list:
    with open(file_path, 'r', encoding='utf-8') as file:
        markdown_content = file.read()  # íŒŒì¼ì˜ ëª¨ë“  ë‚´ìš©ì„ ì½ìŒ
    lines = [i for i in markdown_content.strip().split("\n")  if i != ""]
    data = {
        "function": lines[0].strip('# ').strip(),  # ì²« ë²ˆì§¸ ì¤„ì—ì„œ ## ë¥¼ ì œê±°í•˜ê³  ê³µë°± ì œê±°
        "description": lines[1].strip(),       # ë‘ ë²ˆì§¸ ì¤„ì˜ ë‚´ìš©
        "contributor": lines[2].strip('By ').strip(),  # ì„¸ ë²ˆì§¸ ì¤„ì—ì„œ Byë¥¼ ì œê±°í•˜ê³  ê³µë°± ì œê±°
        "reference": lines[3].strip(),               # ë„¤ ë²ˆì§¸ ì¤„ì˜ ë‚´ìš©
        "prompt" : "\n".join(lines[5:-1])
    }
    result.append(data)
```
